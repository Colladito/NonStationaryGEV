# Functional example of the Automatic non stationary GEV 

In this example, it is shown a basic usage of the Automatic non stationary GEV class to select and calculate the parameters which minimize the AIC related to time-dependent GEV distribution using the Maximum Likelihood method within an iterative scheme, including one parameter at a time based on a perturbation criteria. The process is repeated until no further improvement in the objective function is achieved.


```python
import pandas as pd
import numpy as np
from Automatic_GEV_fit import GEVnonstat
```

```python
hsmax = np.loadtxt("HsmaxBretagne.txt")
datemax = np.loadtxt("datemaxBretagne.txt")
datemax_annual = datemax % 1
indices = np.loadtxt("covBretagne.txt")
indices_dict = {i: indices[:, i] for i in range(indices.shape[1])}
```

```python
autogev = GEVnonstat(xt = hsmax, t = datemax, kt = None, covariates=indices_dict, trends=True, ejemplo="Bretagne")
beta0,beta,alpha0,alpha,gamma0,gamma,betaT,auxvarphi_loc,list_loc,betaT2,auxvarphi_sc,list_sc,loglikeobj,grad,hessian,popt,stdpara = autogev.AutoAdjust()
```

```python
print("beta0",beta0)
print("beta",beta)
print("alpha0",alpha0)
print("alpha",alpha)
print("gamma0",gamma0)
print("gamma",gamma)
print("betaT",betaT)
print("varphi",auxvarphi_loc)
print("betaT2",betaT2)
print("varphi2",auxvarphi_sc)
```
